```python
# change column timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

df.info()
df.head()
```

```python
# check missing values
print(df.isnull().sum())
```

```python
# Create new column is_late. If delivery is more than 3 hours it will be considered late
df["is_late"] = df["delivery_time_deviation"] > 3
```

```python
# Detect outlier(Extreme Values)
print(df["shipping_costs"].describe()) 
print(df["fuel_consumption_rate"].describe())
print(df["delivery_time_deviation"].describe())
```

```python
# Function to detect and handle outliers using IQR method
def handle_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Filter out the outliers
        df[column] = df[column].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
    
    return df

# List of columns to check for outliers
outlier_columns = [
    'fuel_consumption_rate', 'traffic_congestion_level', 'warehouse_inventory_level', 
    'loading_unloading_time', 'handling_equipment_availability', 'order_fulfillment_status', 
    'weather_condition_severity', 'port_congestion_level', 'shipping_costs', 
    'supplier_reliability_score', 'lead_time_days', 'historical_demand', 
    'iot_temperature', 'cargo_condition_status', 'route_risk_level', 
    'customs_clearance_time', 'driver_behavior_score', 'fatigue_monitoring_score', 
    'disruption_likelihood_score', 'delay_probability', 'delivery_time_deviation'
]

# Handle outliers in the specified columns
df_cleaned = handle_outliers_iqr(df, outlier_columns)

# Display the first few rows after handling outliers
df_cleaned.head()
```

```python
# Create categorize from traffic_congestion_level
def categorize_congstion(val):
    if val < 4:
        return "Low"
    elif val < 7:
        return "Medium"
    else :
        return "High"


df["congestion_category"] = df["traffic_congestion_level"].apply(categorize_congstion)
df.head()
```

```python
# Breakdown time to hour, day, weekday, and month
df["hour"] = df["timestamp"].dt.hour
df["day"] = df["timestamp"].dt.day
df["weekday"] = df["timestamp"].dt.day_name()
df["month"] = df["timestamp"].dt.month

df.tail()
```

```python
# Mapping risk_classification to numeric(For the purposes of correlation/statistical analysis or modeling)
risk_map = {'Low Risk': 0, 'Moderate Risk': 1, 'High Risk': 2}
df['risk_classification_code'] = df['risk_classification'].map(risk_map)
```

```python
# Correctly map the 'order_fulfillment_status' based on value thresholds
df_cleaned['order_fulfillment_status_category'] = df_cleaned['order_fulfillment_status'].apply(lambda x: 'Fulfilled' if x > 0.5 else 'Not Fulfilled')

# Show a sample of the data after the transformation
df_cleaned.head()
```
